```python
def preanal(df):
  df_new = pd.DataFrame(columns = df.columns[1:], index=['mean', 'std', 'mean+2sd', 'mean-2sd', 'mean_cleaned'])
  df_new_cleaned = pd.DataFrame()
  for (columnName, columnData) in df.iteritems():
    if 'Trial' in columnName:
      continue
    df_new.loc['mean',columnName] = columnData.mean()
    df_new.loc['std',columnName] = columnData.std()
    trsh1 = columnData.mean() - 2*columnData.std()
    trsh2 = columnData.mean() + 2*columnData.std()
    df_new.loc['mean+2sd',columnName] = trsh1
    df_new.loc['mean-2sd',columnName] = trsh2
    if trsh1 < 0:
      trsh1 = 0
    for j in range(len(df)):
        if (float(df.loc[j, columnName]) >= trsh1) and (float(df.loc[j, columnName]) <= trsh2):
            df_new_cleaned.at[j, columnName + '_' + 'mean+-2sd'] = df.loc[j, columnName]
        else:
            df_new_cleaned.at[j, columnName + '_' + 'mean+-2sd'] = None
    df_new.loc['mean_cleaned',columnName] = df_new_cleaned[columnName + '_' + 'mean+-2sd'].mean()
  return df_new, df_new_cleaned
```
